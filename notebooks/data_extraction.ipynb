```python
stock = yf.Ticker("RELIANCE.NS")
Prices = stock.history(period="5y")
print("Prices shape:", Prices.shape)
print(Prices.head(3))

income = stock.financials.T
print(income.head(3))

balance=stock.balance_sheet.T
print(balance.head(3))

cashflow=stock.cashflow.T
print(cashflow.head(3))

info = stock.info
ratios = {
    "Market Cap" : info.get("marketCap"),
    "P/E Ratio" : info.get("trailingPE"),
   "price to book" : info.get("priceToBook"),
    "Dividend Yield" : info.get("DividendYield"),
    "Return on equity" : info.get("returnOnEquity"),
    "Debt to equity" : info.get("debtToEquity"),
  "Revenue growth" : info.get("revenueGrowth"),

}

ratios_df=pd.DataFrame(ratios, index = [0])
print(ratios_df)

Prices.shape
```

```python
income.to_csv("reliance_income_statement.csv")
balance.to_csv("reliance_balance_sheet.csv")
cashflow.to_csv("reliance_cashflow.csv")
ratios_df.to_csv("reliance_ratios.csv")
Prices.to_csv("reliance_prices.csv")
```
