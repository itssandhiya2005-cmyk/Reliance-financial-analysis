```python
import pandas as pd
import json
import google.generativeai as genai

financial = pd.read_csv("/content/reliance_income_cleaned.csv")
balance = pd.read_csv("/content/reliance_balance_clean_v2.csv")
cashflow = pd.read_csv("/content/reliance_cashflow_cleaned.csv")
income_v2 = pd.read_csv("/content/reliance_income_clean_v2.csv")
ratios = pd.read_csv("/content/reliance_ratios.csv")
ccc= pd.read_csv("/content/reliance ccc.csv")
def to_crores(value):
  try:
    return round(float(value)/100000000,2)
  except:
    return value

financial_summary = []
for _, row in financial.iterrows():
  financial_summary.append({
      "Year": row["Year"],
      "Total_Revenue_crores": to_crores(row["Total Revenue"]),
      "Net_Income_Crores": to_crores(row["Net Income Common Stockholders"]),
      "Operating_Income_Crores" : to_crores(row["Operating Income"]),
      "EBITDA_Crores": to_crores(row["EBITDA"]),
      "Tax_Provision_Crores": to_crores(row["Tax Provision"])
  })
```

```python
balance_summary = []
for _, row in balance.iterrows():
  balance_summary.append({
      "Year": row["Year"],
      "Total_Assets_crores": to_crores(row["Total Assets"]),
      "Total_Debt_Crores": to_crores(row["Total Debt"]),
      "Equity_Crores": to_crores(row["Stockholders Equity"]),
      "Cash_Crores": to_crores(row["Cash And Cash Equivalents"])
  })
```

```python
cashflow_summary = []
for _, row in cashflow.iterrows():
  cashflow_summary.append({
      "Year": row["Year"],
      "Operating_CF_crores": to_crores(row["Operating Cash Flow"]),
      "Free_CF_crores": to_crores(row["Free Cash Flow"]),
      "Investing_CF_crores": to_crores(row["Investing Cash Flow"])
  })
```

```python
ccc_summary = []
for _, row in ccc.iterrows():
  ccc_summary.append({
      "Year": row["Year"],
      "Days_Inventory_Outstanding": row["Days_Inventory_Outstanding"],
      "Days_Sales_Outstanding": row["Days_Sales_Outstanding"],
      "Days_Payables_Outstanding": row["Days_Payable_Outstanding"],
      "Cash_Conversion_Cycle_Days": row["Cash_Conversion_Cycle"],
      })
```

```python
ratios_summary = []
for _, row in ratios.iterrows():
  ratios_summary = {
       "PE_Ratio": row["P/E Ratio"],
        "Price_to_Book": row["price to book"],
        "Return_on_Equity": row["Return on equity"],
        "Debt_to_Equity": row["Debt to equity"],
        "Revenue_Growth": row["Revenue growth"]
  }
```

```python
all_findings = {
    "company": "Reliance Industries Limited",
    "income_statement": financial_summary,
    "balance_sheet": balance_summary,
    "cash_flow": cashflow_summary,
    "ratios": ratios_summary,
    "cash_conversion_cycle": ccc_summary,
    "valuation_ratios": ratios,
}
```

```python
findings_text = json.dumps(all_findings,indent=2,default=str)
```

```python

prompt = f"""
You are a senior financial analyst at a top-tier management consulting firm
like Deloitte. A client has requested an executive briefing on Reliance
Industries Limited based on a detailed SQL-driven financial analysis
covering FY2023 to FY2026.

Write a professional executive briefing with exactly 4 paragraphs:

Paragraph 1 — Revenue & Profitability:
Summarize the revenue growth trend and what accelerating growth
(2.65% to 7.06% to 9.59%) signals about business momentum.
Mention profit margin stability around 7.2% to 7.7%.

Paragraph 2 — Financial Health & Risk:
Assess balance sheet strength, debt position (D/E consistently ~0.44),
and cash reserves growing nearly 4x from ₹34,156 Cr to ₹1,37,327 Cr.
Give a clear confident verdict on financial safety.

Paragraph 3 — Cash Efficiency (Most Important Paragraph):
Analyze the Cash Conversion Cycle trend carefully.
Highlight that FY2026 showed a sharp deterioration to 24.0 days
despite being the strongest revenue year.
Explain that receivable days nearly doubled (11.8 to 20.2) while
payable days shortened (94.3 to 73.7) creating hidden liquidity
pressure a client would completely miss if only looking at
headline revenue numbers.
Explain why this matters for investment or partnership decisions.

Paragraph 4 — Forward Outlook & Recommendation:
Based on accelerating revenue, strong balance sheet, FCF turnaround
from -₹25,956 Cr (FY2023) to +₹69,197 Cr (FY2026), and the CCC
warning signal — give one clear specific actionable recommendation
a Deloitte consultant would make to a client.

Writing rules:
- Formal but clear consulting language
- Each paragraph exactly 4-5 sentences
- No bullet points — flowing prose only
- Use specific numbers from the data
- This will be presented directly to a senior client

Financial Data (imported directly from SQL database via Python):
{findings_text}
"""

response = model.generate_content(

    contents=prompt
)

print("=" * 70)
print("RELIANCE INDUSTRIES LIMITED")
print("AI-GENERATED EXECUTIVE BRIEFING")
print("Model: Google Gemini 2.5 Flash")
print("Data: Imported from MySQL exports via Python (FY2023-FY2026)")
print("=" * 70)
print()
print(response.text)
```
